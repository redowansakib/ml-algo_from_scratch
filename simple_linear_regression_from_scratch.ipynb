### Method 
Here we will be computing simple linear regression for one input variable and one output variable. The one coefficeint will be computed by dividing the covariance of input and output variable by variance of x. Intercept will be computed by subtracting the multiplication of coefficient and mean of input varaible from mean of output variable.

```python
def simple_linear_regression(data):
    x = data[:,0]
    y = data[:,1] 
    coef = np.cov(x,y)[0,1]/np.var(x)
    intercept = mean(y) - coef*mean(x)
    return coef, intercept 
```

```python
coef, intercept = simple_linear_regression(data)
coef, intercept
```
